#### Import Data into Pandas Dataframe

```python
PATH = 'data/Gossip/'
mat_path = 'RFENetwork.mat'
loaded_mat = spio.loadmat(PATH + mat_path)
adj_mat = loaded_mat['Z'][0]
```

#### Construct Graph

```python
# 70 villages; across 7 dimensions - borrow, lend, coming free time, going free time, seek advice, give advice, relative.
one_village = adj_mat[0]
dimensions = one_village[0]
```

```python
borrow, lend, coming_free_time, going_free_time, seek_advice, give_advice, relative = [], [], [], [], [], [], []
columns = [borrow, lend, coming_free_time, going_free_time, seek_advice, give_advice, relative]

for i in range(len(dimensions)):
    cur_dim = dimensions[i]
    for j in range(len(columns)):
        print(cur_dim[j])
```
